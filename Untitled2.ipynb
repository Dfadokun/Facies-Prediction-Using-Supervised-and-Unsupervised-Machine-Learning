```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context()


%matplotlib inline
```

```python
AW3= pd.read_csv('ADSW_4ST3.csv')

AW3
```

```python
AW3=AW3.apply(pd.to_numeric,errors="ignore")
AW3
```

```python
AW3 = AW3.replace(["ADSW-4ST3",'-9999'],['AW3',np.nan])
AW3 = AW3.drop(['datasetName'],axis=1)
AW3 = AW3.drop([0],axis=0)
AW3 = AW3.rename({"DEPT":'Depth'},axis='columns')
AW3
```

```python
AW3=AW3.reset_index()
AW3=AW3.drop(['index'],axis=1)
AW3=AW3.drop(['PHIE_D',"ILD"],axis=1)
AW3=AW3.dropna()

AW3
```
