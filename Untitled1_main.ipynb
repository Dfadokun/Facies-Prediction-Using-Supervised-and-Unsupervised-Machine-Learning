```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context()


%matplotlib inline
```

```python
AW1= pd.read_csv('AD_1.csv')

AW1
```

```python
AW1['GR_NORM']
```

```python
AW1 = AW1.replace(["AD-1",'-9999'],['AW1',np.nan])
AW1 = AW1.drop(['datasetName'],axis=1)
AW1 = AW1.drop([0],axis=0)
AW1 = AW1.rename({"DEPT":'Depth'},axis='columns')
AW1
```

```python
AW1=AW1.reset_index()
AW1=AW1.drop(['index'],axis=1)
AW1
```

```python
AW2=AW1[AW1["Depth"].astype(float)>6900]
AW2['ILD_log10']=np.log10 (AW2["ILD"].astype(float))
AW2 = AW2.drop(["PHIE_D",'ILD','DT'],axis=1)
AW2 = AW2.dropna()

AW2
```

```python
AW2 =AW2.apply(pd.to_numeric,errors="ignore")
AW2
```

```python
AW2.describe()
```

```python
x = AW2.iloc[:,3:7]
x
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method = "ward"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, affinity ='euclidean',linkage ='ward')


AW2['facies'] = hc.fit_predict (x)
label = hc.labels_
AW2
```

```python
from sklearn.cluster import AgglomerativeClustering
hc1 = AgglomerativeClustering(n_clusters=4, affinity ='euclidean',linkage ='ward')


AW2['facies1'] = hc1.fit_predict (x)
label = hc.labels_
AW2
```

```python
plt.style.use('seaborn')
plt.scatter(AW2['PHIT_D'].astype(float), AW2['VSH_FINAL'].astype(float),c=AW2['facies'].astype(float),cmap = 'viridis')
plt.rcParams["figure.figsize"]=[10,10]
plt.ylabel("Volume of Shale")
plt.xlabel("Porosity")
plt.show()
```

```python
plt.style.use('seaborn')
plt.scatter(AW2['PHIT_D'].astype(float), AW2['VSH_FINAL'].astype(float),c=AW2['facies1'].astype(float),cmap = 'viridis')
plt.rcParams["figure.figsize"]=[10,10]
plt.ylabel("Volume of Shale")
plt.xlabel("Porosity")
plt.show()
```

```python
from sklearn import metrics
```

```python
print(metrics.silhouette_score(x,AW2['facies'],metric="euclidean"))
```

```python
print(metrics.silhouette_score(x,AW2['facies1'],metric="euclidean"))
```

```python
Facies=[]
for  i in AW2["facies1"].values:
    if float(i)==0:
        Facies.append(2)
    elif float(i)==2:
        Facies.append(3)
    elif float(i)==1:
        Facies.append(1)
    else:
        Facies.append(0)
AW2["Facies"] = Facies
AW2
```

```python
plt.style.use('seaborn')
fig,(ax1,ax2,ax4) = plt.subplots(figsize = (6,20),nrows=1,ncols=3,sharey=True)

print (ax1)
print (ax2)

ax1.plot(AW2["GR_NORM"].astype(float),AW2["Depth"].astype(float), color = "g", label = 'Gamma Ray')
ax1.set_ylim(8000,6800)
ax1.set_xlim(0,150)
ax1.set_title ("GR")
ax1.legend()
ax2.plot(AW2["RHOB_HC"].astype(float),AW2["Depth"].astype(float), color = "r", label = 'DEN')
ax2.set_ylim(8000,6800)
ax2.set_xlim(1.65,2.65)
ax2.legend()
ax3 = ax2.twiny() 

ax3.plot(AW2["NPHI_HC"].astype(float),AW2["Depth"].astype(float), color = "b", label = "NEU")
ax3.set_ylim(8000,6800)
ax3.set_xlim(0.6,0)
ax3.set_title ("NEU-DEN")
ax3.legend()

ax4.plot(AW2["Facies"].astype(float),AW2["Depth"].astype(float), color = "black", label = "Facies")
ax4.set_ylim(8000,6800)
ax4.set_xlim(0,4)
ax4.set_title ("Facies")
ax4.legend()
fig.savefig("AW1_logplot.png")
```

```python
AW3_1 = AW3.dropna()
x3 = AW3_1.iloc[:,4:9 ]
x3.drop(['PHIE_D'], axis=1, inplace=True)
x3
```
